```
# Importing Pandas
import pandas as pd
# Importing numpy
import numpy as np
```

```
#Loading the Telcom datasets 
#We create a URL to Excel file (alternatively this can be a filepath)
url = '/content/Telcom_dataset.csv'

# Then load the first sheet of the Excel file into a data frame
Telcom_dataframe1 = pd.read_csv(url)

# Lastly, view the first ten rows
Telcom_dataframe1.head(10)
```

```
#Cleaning the data
#Find and remove null/missing values and duplicates
#
#drop duplicates from the dataframe
Telcom1_dropped_duplicates = Telcom_dataframe1.drop_duplicates()
#delete missing values from the dataframe
#
Telcom_dataframe1_no_missing = Telcom1_dropped_duplicates.dropna()
Telcom_dataframe1_no_missing
#delete rows with no data at all
#
Telcom_dataframe1_final = Telcom_dataframe1_no_missing.dropna(how='all')
Telcom_dataframe1_final
```

```
#We create a URL to Excel file (alternatively this can be a filepath)
url2 = '/content/Telcom_dataset2.csv'

# Then load the first sheet of the Excel file into a data frame
Telcom_dataframe2 = pd.read_csv(url2)

# Lastly, view the first ten rows
Telcom_dataframe2.head(10)
```

```
#Cleaning the data
#Find and remove null/missing values and duplicates
#
#drop duplicates from the dataframe
Telcom2_dropped_duplicates = Telcom_dataframe2.drop_duplicates()
#delete missing values from the dataframe
#
Telcom_dataframe2_no_missing = Telcom2_dropped_duplicates.dropna()
Telcom_dataframe2_no_missing
#delete rows with no data at all
#
Telcom_dataframe2_final = Telcom_dataframe2_no_missing.dropna(how='all')
Telcom_dataframe2_final
```

```
#We create a URL to Excel file (alternatively this can be a filepath)
url3 = '/content/Telcom_dataset3.csv'

# Then load the first sheet of the Excel file into a data frame
Telcom_dataframe3 = pd.read_csv(url3)

# Lastly, view the first ten rows
Telcom_dataframe3.head(10)
```

```
#Cleaning the data
#Find and remove null/missing values and duplicates
#
#drop duplicates from the dataframe
Telcom3_dropped_duplicates = Telcom_dataframe3.drop_duplicates()
#delete missing values from the dataframe
#
Telcom_dataframe3_no_missing = Telcom3_dropped_duplicates.dropna()
Telcom_dataframe3_no_missing
#delete rows with no data at all
#
Telcom_dataframe3_no_data = Telcom_dataframe3_no_missing.dropna(how='all')
Telcom_dataframe3_no_data
#rename the 'SIET_ID' column 
#
Telcom_dataframe3_final = Telcom_dataframe3_no_data.rename(columns = {'SIET_ID':'SITE_ID'}, inplace=False)
Telcom_dataframe3_final
```

```
#We create a URL to Excel file (alternatively this can be a filepath)
url4 = '/content/cells_geo.csv'

# Then load the first sheet of the Excel file into a data frame
Geo_dataframe1 = pd.read_csv(url4)

# Lastly, view the first ten rows
Geo_dataframe1.head(10)
```

```
#Cleaning the data
#Find and remove null/missing values and duplicates
#
#drop duplicates from the dataframe
Geo_dataframe_drop_duplicates = Geo_dataframe1.drop_duplicates()
#delete missing values from the dataframe
#
Geo_dataframe_no_missing = Geo_dataframe_drop_duplicates.dropna()
Geo_dataframe_no_missing
#delete rows with no data at all
#
Geo_dataframe_no_data = Geo_dataframe_no_missing.dropna(how='all')
Geo_dataframe_no_data
#rename the 'SITE_CODE' column 
#
Geo_dataframe_final = Geo_dataframe_no_data.rename(columns = {'SITE_CODE':'SITE_ID'}, inplace=False)
Geo_dataframe_final
```

```
#Merge the dataframes into a single dataset for analysis
#
Final_Telcom_dataframe = Telcom_dataframe1_final.merge(Telcom_dataframe2_final,on='SITE_ID').merge(Telcom_dataframe3_final,on='SITE_ID').merge(Geo_dataframe_final,on='SITE_ID')
Final_Telcom_dataframe
```

```
# Create a groupby variable that groups product by Region 
#This will allow us to see what MTN product is most used in certain regions 
#
Telcom_groupby= Final_Telcom_dataframe['PRODUTC'].groupby(Final_Telcom_dataframe['REGION'])
Telcom_groupby
```

```
#This shows us the region tallies and top products in those regions.
#Voice services is the highest product for majority of the regions, while only a few use sms as the top product
#Infrastructure should be biased towards voice-centered technologies 
# 
Telcom_groupby.describe()
```
